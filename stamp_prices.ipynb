# Plot US Stamp Prices Over Time

```python
# Import necessary libraries (pandas & matplotlib)
import pandas as pd
import matplotlib.pyplot as plt
```

This dataset is from https://vincentarelbundock.github.io/Rdatasets/datasets.html

```python
# I started with pd.read_csv to turn the csv into a pandas dataframe and then confirmed the conversion with type()
stamp_prices = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/Stat2Data/USstamps.csv')

type(stamp_prices)
```

```python
# head() returns the column names and first few rows of our new dataframe stamp_prices. Note that the price is in cents. It's not listed here but is stated to be in cents in the DOC for the data set.
stamp_prices.head()
```

```python
# Plotting the year vs the price. The first step is to set a plot title.
plt.title('Stamp Prices (in Cents) Over Time in the US')

#Set the columns to plot, x-axis to be listed first then y-axis
plt.scatter(stamp_prices['Year'], stamp_prices['Price'])
```
